- 첫 번째 : 효율성테스트 3, 4에서 실패
    - sort()를 모든 query에 하기 때문에 비효율이 일어남

```python
# 통과 (6696.93ms, 42.5MB)
# 통과 (6415.11ms, 43MB)
from itertools import product
from bisect import bisect_left

def solution(info, query):        
    dic = dict()
    condition = [['cpp', 'java', 'python', '-'], ['backend', 'frontend', '-'], 
                  ['junior', 'senior', '-'], ['chicken', 'pizza', '-']]
    conditions = list(product(*condition))
    for i in conditions:
        dic[''.join(i)] = list()
    
    for i in info:
        temp = i.split()
        score = int(temp[-1])
        temp = temp[:-1]
        dic[''.join(temp)].append(score)
        dic['-' + temp[1] + temp[2] + temp[3]].append(score)
        dic[temp[0] + '-' + temp[2] + temp[3]].append(score)
        dic[temp[0] + temp[1] + '-' + temp[3]].append(score)
        dic[temp[0] + temp[1] + temp[2] + '-'].append(score)
        dic['-' + '-' + temp[2] + temp[3]].append(score)
        dic['-' + temp[1] + '-' + temp[3]].append(score)
        dic['-' + temp[1] + temp[2] + '-'].append(score)
        dic[temp[0] + '-' + '-' + temp[3]].append(score)
        dic[temp[0] + '-' + temp[2] + '-'].append(score)
        dic[temp[0] + temp[1] + '-' + '-'].append(score)
        dic['-' + '-' + '-' + temp[3]].append(score)
        dic['-' + '-' + temp[2] + '-'].append(score)
        dic['-' + temp[1] + '-' + '-'].append(score)
        dic[temp[0] + '-' + '-' + '-'].append(score)
        dic['-' + '-' + '-' + '-'].append(score)
    
    answer = list()
    for i in query:
        temp = i.split()
        score = int(temp[-1])
        scores = dic[temp[0] + temp[2] + temp[4] + temp[6]]
        scores.sort()
        idx = bisect_left(scores, score)
        answer.append(len(scores[idx:]))
        
    return answer
```

- 세 번째 : 통과

```python
from itertools import product
from bisect import bisect_left

def solution(info, query):        
    dic = dict()
    condition = [['cpp', 'java', 'python', '-'], ['backend', 'frontend', '-'], 
                  ['junior', 'senior', '-'], ['chicken', 'pizza', '-']]
    conditions = list(product(*condition))
    for i in conditions:
        dic[''.join(i)] = list()
    
    for i in info:
        temp = i.split()
        score = int(temp[-1])
        temp = temp[:-1]
        dic[''.join(temp)].append(score)
        dic['-' + temp[1] + temp[2] + temp[3]].append(score)
        dic[temp[0] + '-' + temp[2] + temp[3]].append(score)
        dic[temp[0] + temp[1] + '-' + temp[3]].append(score)
        dic[temp[0] + temp[1] + temp[2] + '-'].append(score)
        dic['-' + '-' + temp[2] + temp[3]].append(score)
        dic['-' + temp[1] + '-' + temp[3]].append(score)
        dic['-' + temp[1] + temp[2] + '-'].append(score)
        dic[temp[0] + '-' + '-' + temp[3]].append(score)
        dic[temp[0] + '-' + temp[2] + '-'].append(score)
        dic[temp[0] + temp[1] + '-' + '-'].append(score)
        dic['-' + '-' + '-' + temp[3]].append(score)
        dic['-' + '-' + temp[2] + '-'].append(score)
        dic['-' + temp[1] + '-' + '-'].append(score)
        dic[temp[0] + '-' + '-' + '-'].append(score)
        dic['-' + '-' + '-' + '-'].append(score)
    
    for value in dic.values():           
        value.sort()
    
    answer = list()
    for i in query:
        temp = i.split()
        scores = temp[0] + temp[2] + temp[4] + temp[6]
        if scores not in dic:
            continue
        scores = dic[scores]
        score = int(temp[-1])
        idx = bisect_left(scores, score)
        answer.append(len(scores[idx:]))
        
    return answer
```
